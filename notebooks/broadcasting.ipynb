```python
# Import
import numpy as np
```

```python
# Broadcasting with Scalars

# Create array
arr = np.array([1, 2, 3, 4, 5])

# Scalar addition
result = arr + 10

print("Original:", arr)
print("After adding scalar:", result)

# A scalar is applied to every element of the array.
```

```python
#Broadcasting Between 1D Arrays

a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

print("a:", a)
print("b:", b)

print("Addition:", a + b)

# Arrays must have same length or compatible shapes.
```

```python
#Broadcasting (2D + 1D)

# 2D array
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

# 1D array
row = np.array([10, 20, 30])

print("Matrix:\n", matrix)
print("Row:", row)

print("Result:\n", matrix + row)
```

```python
# Shape understanding

print("Matrix shape:", matrix.shape)
print("Row shape:", row.shape)

# NumPy aligns shapes from right to left.
```

```python
# Column Broadcasting 

col = np.array([[10], [20]])

print("Column:\n", col)
print("Result:\n", matrix + col)
```

Final Conclusion

Broadcasting allows NumPy to perform operations on arrays of different shapes.
It works by expanding smaller arrays without copying data.
Understanding shapes helps avoid errors and write efficient code.
