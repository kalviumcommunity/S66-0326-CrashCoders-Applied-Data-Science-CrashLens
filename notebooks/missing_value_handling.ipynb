```python
#import and load
import pandas as pd

df = pd.read_csv("../data/raw/sample_data.csv")

print("Original Data:")
print(df)
print("Shape:", df.shape)
```

```python
#Dropping missing values

#Drop rows with any missing values
df_dropped = df.dropna()

print("After dropping rows:")
print(df_dropped)
print("Shape after drop:", df_dropped.shape)

#Drop columns with any missing values
df_drop_col = df.dropna(axis=1)

print("After dropping columns:")
print(df_drop_col)
```

Dropping removes missing data but reduces dataset size.

```python
#Filling missing values

#Fill with constant
df_fill_const = df.fillna("Unknown")

print("After filling with constant:")
print(df_fill_const)

#Fill numeric values with mean
df_fill_mean = df.copy()

df_fill_mean["age"] = df_fill_mean["age"].fillna(df_fill_mean["age"].mean())

print("After filling age with mean:")
print(df_fill_mean)
```

Filling preserves data but introduces assumptions.

```python
#Compare Results
print("Original shape:", df.shape)
print("After drop:", df_dropped.shape)
print("After fill:", df_fill_mean.shape)
```

Dropping reduces rows, while filling keeps dataset size same.

#Final Conclusion

Missing values can be handled by dropping or filling.
Dropping removes incomplete data but reduces dataset size.
Filling preserves data but requires careful assumptions.
Choosing the right method depends on context.
