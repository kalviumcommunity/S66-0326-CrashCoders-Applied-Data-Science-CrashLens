```python
#Import
import pandas as pd
```

```python
#DataFrame from Dictionary
# Create dictionary
data = {
    "name": ["Khushil", "Yashasvi", "Fiba"],
    "age": [20, 21, 22],
    "city": ["Delhi", "Mumbai", "Bangalore"]
}

# Convert to DataFrame
df_dict = pd.DataFrame(data)

print("DataFrame from dictionary:")
print(df_dict)
```

```python
#Inspect Data Frame
print("Columns:", df_dict.columns)
print("Shape:", df_dict.shape)
print("Data types:\n", df_dict.dtypes)
```

```python
#Load DataFrame from CSV
df_file = pd.read_csv("../data/raw/sample_data.csv")

print("DataFrame from file:")
print(df_file)
```

```python
#Inspect File Data
print("First 2 rows:\n", df_file.head(2))
print("Columns:", df_file.columns)
print("Shape:", df_file.shape)
```

Conclusion

A DataFrame is a tabular data structure with rows and columns.
It allows structured data handling similar to spreadsheets.
DataFrames can be created from dictionaries or loaded from files.
They are essential for data analysis in Pandas.
