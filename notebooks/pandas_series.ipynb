```python
#import

import pandas as pd
import numpy as np
```

```python
#Series from list

# Create list
data_list = [10, 20, 30, 40]

# Convert to Series
series_list = pd.Series(data_list)

print("Series from list:")
print(series_list)
```

Pandas automatically assigns index starting from 0.

```python
#Series from NumPy Array

# Create NumPy array
arr = np.array([1, 2, 3, 4])

# Convert to Series
series_array = pd.Series(arr)

print("Series from NumPy array:")
print(series_array)
```

```python
#Custom Index

series_custom = pd.Series(data_list, index=['a', 'b', 'c', 'd'])

print("Series with custom index:")
print(series_custom)
```

```python
#Indexes and Values

print("Index:", series_custom.index)
print("Values:", series_custom.values)
```

```python
#Accessing Data

# By index label
print("Value at index 'a':", series_custom['a'])

# By position
print("Value at position 0:", series_custom.iloc[0])
```

```python
#Simple Operations

# Add 5 to all elements
print(series_list + 5)

# Multiply
print(series_list * 2)
```

Conclusion

A Pandas Series is a labeled one-dimensional data structure.
It adds index labels to data, making it more meaningful than NumPy arrays.
Series are the foundation for working with DataFrames.
