```python
#import and load

import pandas as pd

df = pd.read_csv("../data/raw/sample_data.csv")

print(df)
```

```python
#Detect Missing Values
print("Missing values (True = missing):")
print(df.isnull())
```

isnull() returns True where values are missing.

```python
#Count Missing Values
print("Missing values per column:")
print(df.isnull().sum())
```

sum() counts total missing values in each column.   

```python
#Check Any Missing Values
print("Any missing values in dataset:", df.isnull().values.any())
```

```python
#Rows with Missing Values
print("Rows with missing values:")
print(df[df.isnull().any(axis=1)])
```

any(axis=1) finds rows that contain at least one missing value.

#Final Conclusion

Missing values are common in datasets.
isnull() helps detect them, and sum() helps count them.
Identifying missing data early prevents errors in analysis.
