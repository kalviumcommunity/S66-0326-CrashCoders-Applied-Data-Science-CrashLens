```python
#import and load
import pandas as pd

df = pd.read_csv("../data/raw/sample_data.csv")

print("Data Loaded")
```

```python
#Check shape
print("Shape of DataFrame:", df.shape)
```

Shape returns a tuple (rows, columns).
Rows represent observations, and columns represent features.

```python
#Rows and Columns
rows, cols = df.shape

print("Number of rows:", rows)
print("Number of columns:", cols)
```

Rows represent individual records.
Columns represent attributes or variables.

```python
#Column Data Types
print("Data Types:\n", df.dtypes)
```

Each column has a specific data type such as int, float, or object.
Correct data types are important for performing valid operations.

```python
#Detect Type Issues
print("Checking numeric column (age):", df["age"].dtype)
```

If numeric data is stored as object (string), it can cause issues in calculations.

```python
#Extra Check
print(df.info())
```

Final Conclusion

DataFrame shape tells us the size of the dataset.
Data types define how each column behaves.
Understanding shape and types helps prevent errors in analysis.
