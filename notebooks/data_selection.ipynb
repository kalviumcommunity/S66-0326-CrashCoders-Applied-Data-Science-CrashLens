```python
#import
import pandas as pd

df = pd.read_csv("../data/raw/sample_data.csv")

print("Data Loaded")
print(df)
```

```python
#Selecting Columns

#Single Column
print("Single column (name):")
print(df["name"])

#Multiple Columns
print("Multiple columns:")
print(df[["name", "age"]])
```

Columns can be selected using column names.
Single column → Series
Multiple columns → DataFrame

```python
#Selecting rows by position(iloc)
print("First row:")
print(df.iloc[0])

print("First 2 rows:")
print(df.iloc[0:2])
```

iloc is used for position-based indexing.
It uses integer positions starting from 0.

```python
#Selecting rows by label(loc)
print("Row with index 0:")
print(df.loc[0])
```

loc is used for label-based indexing.
It uses index labels instead of positions.

```python
#Selecting rows and columns together
print("Specific rows and columns:")
print(df.loc[0:1, ["name", "city"]])
```

We can select both rows and columns together using loc.
Format: df.loc[row_selection, column_selection]

#Final Conclusion
Column selection is done using column names.
Row selection can be done using iloc (position) or loc (label).
Combining both allows precise data extraction.
