```python
#import and load
import pandas as pd

df = pd.read_csv("../data/raw/sample_data.csv")

print("Data Loaded")
```

```python
#head()
print("First 5 rows:")
print(df.head())
```

head() shows the first few rows of the dataset.
It helps quickly verify column names and sample data.

```python
#info()
print("DataFrame Info:")
df.info()
```

info() provides column names, data types, and non-null counts.
It helps identify missing values and understand structure.

```python
#describe()
print("Statistical Summary:")
print(df.describe())
```

describe() gives summary statistics for numeric columns.
It includes count, mean, min, max, and percentiles.

```python
#Combined Understanding
print("Columns:", df.columns)
print("Shape:", df.shape)
```

Final Conclusion

head() helps preview data,
info() explains structure and data types,
describe() provides statistical summary.
Together, they help understand the dataset before analysis.
