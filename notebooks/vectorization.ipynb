```python
import numpy as np
```

```python
# Create array
arr = np.array([1, 2, 3, 4, 5])
print("Original Array:", arr)
```

```python
# Loop-based operation
result_loop = []

for i in arr:
    result_loop.append(i * i)

print("Loop Result:", result_loop)
```

```python
# Vectorized operation

result_vectorized = arr * arr
print("Vectorized Result:", result_vectorized)
```

```python
# Arithmetic Operations

print("Add 10:", arr + 10)
print("Multiply by 2:", arr * 2)
```

```python
# Comparisons

print("Greater than 3:", arr > 3)
```

### Conclusion

Vectorized operations are faster and more readable than loops.
They allow operations on entire arrays at once.
NumPy is optimized for vectorization.
