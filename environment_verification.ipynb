## 3. Essential Data Science Packages Verification

```python
print("=" * 60)
print("ESSENTIAL PACKAGES VERIFICATION")
print("=" * 60)

packages_to_test = [
    ('numpy', 'np'),
    ('pandas', 'pd'),
    ('matplotlib', 'plt'),
    ('sklearn', 'sklearn')
]

for package_name, alias in packages_to_test:
    try:
        package = __import__(package_name)
        version = getattr(package, '__version__', 'version unknown')
        print(f"✓ {package_name:15} ({alias:4}): {version}")
    except ImportError as e:
        print(f"✗ {package_name:15} - Not installed")

print("=" * 60)
```

## 4. NumPy Functionality Test

```python
import numpy as np

print("Testing NumPy operations...")
print()

# Create array
arr = np.array([1, 2, 3, 4, 5])
print(f"✓ NumPy array created: {arr}")

# Array operations
print(f"✓ Array mean: {arr.mean()}")
print(f"✓ Array std: {arr.std():.2f}")

# Matrix creation
matrix = np.random.rand(3, 3)
print(f"✓ Random matrix shape: {matrix.shape}")

print()
print("NumPy tests passed! ✓")
```

## 5. Pandas Functionality Test

```python
import pandas as pd

print("Testing Pandas operations...")
print()

# Create DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Score': [85.5, 90.0, 78.5]
})

print("✓ DataFrame created:")
print(df)
print()
print(f"✓ DataFrame shape: {df.shape}")
print(f"✓ Average score: {df['Score'].mean():.2f}")

print()
print("Pandas tests passed! ✓")
```
